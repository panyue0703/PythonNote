```python
import time

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
# url地址
url = 'https://www.baidu.com'
# 定义chrome驱动去地址
path =  Service('C:\Program Files\Google\Chrome\Application\chromedriver.exe')
# 创建浏览器操作对象
web = webdriver.Chrome(service=path)
web.get('https://www.baidu.com/')
```

```python
# 窗口最大化
web.maximize_window()
```

```python
# 获取网页源码
print(web.page_source)
```

```python
# 转到一个网址
web.get('https://cd.zu.ke.com/zufang/pg2/#contentList')
# 回去
web.back()
# 前进
web.forward()
```

## 在浏览器内部进行操作

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

#### 元素定位

```python
search = web.find_element(By.XPATH,'//*[@id="kw"]')     #定位搜索框
```

#### 写入内容

```python
search.send_keys('成都大学')
```

#### 敲击回车键

```python
search.send_keys(Keys.ENTER)
```

#### 获取“百度一下”按钮

```python
baidu = web.find_element(By.XPATH,'//*[@id="su"]')
baidu.click()
time.sleep(2)
```

#### 拿源码

```python
print(web.page_source)
```
