```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
# url地址
url = 'https://global.jd.com/'
# 定义chrome驱动去地址
path =  Service('C:\Program Files\Google\Chrome\Application\chromedriver.exe')
# 创建浏览器操作对象
web = webdriver.Chrome(service=path)
web.get(url)
```

```python
search = web.find_element(By.XPATH,'//*[@id="key"]')
search.send_keys('手机')
search.send_keys(Keys.ENTER)
```

#### 自动下拉操作
获取动态页面数据

```python
max_y = 10000
y = 0
while y <= max_y:
    web.execute_script(f'window.scrollTo(0,{y})')
    y+=1000
    time.sleep(3)
```

获取网页源码

```python
print(web.page_source)
```

#### 定位标签

```python
title =  web.find_element(By.XPATH,'//*[@id="J_goodsList"]/ul/li[1]/div/div[4]/a/em').text
print(title)
```

#### 取src图片

```python
pic = web.find_element(By.XPATH,'//*[@id="J_goodsList"]/ul/li[1]/div/div[1]/a/img').get_attribute('src')
print(pic)
```
